#  BoxPlot - GHS e Hellinger
------
## Definindo distribuição 

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy
import ghsw
import distanceMetrics

import warnings
warnings.filterwarnings('ignore')
#ignora o warning sem sentido dohistograma normalizado!

class betaDistribution:

    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def getDistribution(self,size):
        data = numpy.random.beta(a=self.a,b=self.b,size=[size])
        data += self.c
        return data
    
#exemplos:
b1 = betaDistribution(0.6, 3.0, 0.0)
b2 = betaDistribution(3.0, 0.6, 0.0)
```

## Função para uma densidade

```python
def getBootstrap(beta1, beta2, density, bootstrapSampleSize, weightType=3):
    ghsList = []  
    hellList = []
    KLList = []
    for i in range(bootstrapSampleSize):
        bca,bcl,ghs = ghsw.ghsw(beta1.getDistribution(density), beta2.getDistribution(density), weightType,onlyGHS=False)
        ghsList.append(ghs)
        hellList.append(distanceMetrics.hellinger(beta1.getDistribution(density), beta2.getDistribution(density)))
        KLList.append(distanceMetrics.dkl(beta1.getDistribution(density), beta2.getDistribution(density)))
    return ghsList,hellList,KLList


#testando....
ghs, _, _ = getBootstrap(b1,b2,100,3000)
plt.boxplot(ghs, sym='')
plt.show()
```

## Loop principal

```python
def plotSampleDensity(b1,b2,density):
        plt.figure(figsize=(5,5))
        plt.title("Sample density:"+str(density))
        plt.hist(b1.getDistribution(density),histtype='step',normed=False, color='r',linewidth=1.7)
        plt.hist(b2.getDistribution(density),histtype='step',normed=False, color='b',linewidth=1.7)
        plt.savefig("ex"+str(density)+".png")
    
```

```python
densidades = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
#densidades = [100,200,1000]

lstGHS = []
lstHell = []
lstKull = []

for densidade in densidades:
    ghs, hell, kull = getBootstrap(b1,b2,densidade,3000)
    lstGHS.append(ghs)
    lstHell.append(hell)
    lstKull.append(kull)

#Primeiro plot: GHS
plt.figure(figsize=(10,5))
plt.boxplot(lstGHS, sym='')
plt.xticks(numpy.arange(len(densidades))+1,densidades, rotation=45)
plt.ylabel(r"$\delta_{GHS}(w_{3})$", fontsize=17)

#Segundo plot: Hellinger
plt.figure(figsize=(10,5))
plt.subplots_adjust(hspace = 0.25,wspace= 0.3)

plt.boxplot(lstHell, sym='')
plt.xticks(numpy.arange(len(densidades))+1,densidades, rotation=45)
plt.xlabel("Distribution Density", fontsize=15)
plt.ylabel(r"$D_H$", fontsize=17)

#Terceiro plot: Kullback-Leibler
plt.figure(figsize=(10,5))
plt.subplots_adjust(hspace = 0.25,wspace= 0.3)

plt.boxplot(lstKull, sym='')
plt.xticks(numpy.arange(len(densidades))+1,densidades, rotation=45)
plt.xlabel("Distribution Density", fontsize=15)
plt.ylabel(r"$D_K$", fontsize=17)


#Exemplos:
for i in densidades:
    plotSampleDensity(b1,b2,i)

plt.show()
```
