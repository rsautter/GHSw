```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fakeSampler(prob, xs,nsample=1000):
    normalized = prob.copy()/np.sum(prob)
    output = []
    delta = np.min(np.gradient(xs)/2)
    for idx,p in enumerate(normalized):
        for i in range(int(nsample*p)):
            amostra = xs[idx]+delta*np.random.rand()
            output.append(amostra)
    return output
```

```python
plt.figure(figsize=(5,5))

prob = [0.003,0.027,0.082,0.12,0.059,0.051,0.02,0.005,0.0025]
xs = np.arange(2.5,25,2.5)
sampleQuark  = fakeSampler(prob,xs,10000)
np.savetxt("samples/nQuark.csv",sampleQuark,delimiter=',')

plt.hist(sampleQuark,bins=np.arange(2.5,27.5,2.5)-1.25,histtype='step',lw=2,density=True)

prob = [0.00,0.005,0.02,0.059,0.067,0.11,0.07,0.023,0.0075,0.0001]
xs = np.arange(2.5,27.5,2.5)

sampleGluon  = fakeSampler(prob,xs,10000)
np.savetxt("samples/nGluon.csv",sampleGluon,delimiter=',')
plt.hist(sampleGluon,bins=np.arange(2.5,27.5,2.5)-1.25,histtype='step',color='red',lw=2,ls='--',density=True)

plt.xlabel("n")
plt.ylabel("a.u.")
plt.xlim(0,30)
plt.show()
```
